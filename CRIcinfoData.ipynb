```python
df_data = pd.read_csv('df_data.csv',header=None,skiprows = 1)
df_data2 = pd.read_csv('df_data2.csv',header=None,skiprows = 1)
df_data2.columns = col1
```

```python
df_data_2 = df_data2[df_data2[0].isin(['T20 debut'])]
df_data_2.reset_index(drop=True,inplace=True)
result = [re.findall('[A-Z][^A-Z]*',x) for x in  df_data_2[1]]
debut20 = pd.DataFrame(result)
debut20['T20_debut_1'] = debut20[0] + ' ' +debut20[1]
debut20['T20_debut_2'] = debut20[2] + ' ' + debut20[3]
debut20['T20_debut_2'].fillna('None',inplace=True)
debut20 = debut20[['T20_debut_1','T20_debut_2']]
df_debut20 = pd.concat([debut20,df_data_2['counter']],axis=1)

Replacementdump = df_data2[df_data2[0].str.endswith('Replacement')]


Replacementdump.head()

impact_team = Replacementdump.pivot(values=0,  columns = 'counter').T
dict = {
        "DC Player Replacement" : 'Delhi Capitals', 
        "LSG Player Replacement" : 'Lucknow Super Giants',
        "GT Player Replacement": 'Gujarat Titans',
        "KKR Player Replacement": 'Kolkata Knight Riders',
        "PBKS Player Replacement" : 'Punjab Kings', 
        "RR Player Replacement": 'Rajasthan Royals',
        "SRH Player Replacement": 'Sunrisers Hyderabad',
        "MI Player Replacement": 'Mumbai Indians',
        "RCB Player Replacement": 'Royal Challengers Bangalore',
        "CSK Player Replacement": 'Chennai Super Kings'
         }


impact_team.replace({6: dict,7:dict},inplace=True)
impact_team.reset_index(inplace=True)
impact_team.columns = ['counter','impactTeam1','impactTeam2']

Replacement = Replacementdump.pivot(values=1,  columns = 'counter').T

Replacement.reset_index(inplace=True)

Replacement.columns = ['counter','ImpactTeam1','ImpactTeam2']

def impact(df):
    df1 = df.str.split(":",expand=True)
    df2 = df1[1].str.split(r'in\s*\,',expand=True)
    df3 = df2[1].str.split(r'out\s*\(',expand=True)
    df_final = pd.concat([df2[0],df3[0]],axis=1)
    df_final.fillna('None',inplace=True)
    return df_final

impactTeam1 = impact(Replacement['ImpactTeam1'])
impactTeam1.columns = ['ImpactPlayers1','Replace1Player1']
impactTeam1[impactTeam1.columns] = impactTeam1.apply(lambda x:x.str.strip())


impactTeam2 = impact(Replacement['ImpactTeam2'])
impactTeam2.columns = ['ImpactPlayers2','ReplacePlayer2']
impactTeam2[impactTeam2.columns] = impactTeam2.apply(lambda x:x.str.strip())


df_impact = pd.concat([impactTeam1,impactTeam2,impact_team],axis=1)
df_impact.fillna('None',inplace=True)
```

```python
def clean_batting(batdump):
    batdump= batdump.iloc[:,[0,1,2,3,4,5,6,7,10]].dropna()
    batdump = batdump[(~batdump['BATTING'].str.startswith('Did not bat')) & (~batdump['BATTING'].str.startswith('Fall of wickets'))]   
    batdump.columns = ['batting','out','runs','balls','mins','fours','sixs','SR','counter']
    batdump.loc[batdump['mins'].str.contains('-'),'mins'] = 0
    batdump.loc[batdump['SR'].str.contains('-'),'SR'] = 0

    return batdump

def clean_bowling(dfs):
    dfs['R'] = pd.to_numeric(dfs['R'], errors='coerce')
    dfs = dfs.dropna()
    dfs.columns = ['BOWLING','over','Mins','Runs','Wicket','Economy','Zeros','fours','sixs','Wide','Noball','counter']
    dfs.loc[:, 'Runs'] = dfs.loc[:, 'Runs'].astype('int16')
    return dfs

bat1 = clean_batting(df_bat1dump)
bat2 = clean_batting(df_bat2dump)


bowl1 = clean_bowling(df_bowl1dump)
bowl2 = clean_bowling(df_bowl2dump)

test_join1.to_csv('IPL2023details.csv',header=test_join1.columns, index=False, encoding='utf-8' )
bat1.to_csv('batting1.csv',header=bat1.columns, index=False, encoding='utf-8' )
bat2.to_csv('batting2.csv',header=bat2.columns, index=False, encoding='utf-8' )
bowl1.to_csv('Bowling1.csv',header=bowl1.columns, index=False, encoding='utf-8' )
bowl2.to_csv('Bowling2.csv',header=bowl2.columns, index=False, encoding='utf-8' )
```
